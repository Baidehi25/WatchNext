## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing Dataset

```python
dataset=pd.read_csv('netflix_db.csv',header=None,encoding='latin-1')    #need encoding to read special characters
bucket=[]

for i in range(len(dataset)):
    row=[]
    for j in range(0,20):
        movie=dataset.values[i,j]
        if str(movie)!='nan':
            row.append(str(movie).strip())
    bucket.append(row)
```

## Apriori Implementation

```python
from apyori import apriori
dbrules=apriori(transactions=bucket,min_support=0.003,min_confidence=0.2,min_lift=3,min_length=2)
```

## Visualising Data

### Raw Data

```python
results=list(dbrules)
results
```

### Proper Display (ECLAT model)

```python
def inspect(results):
    movie1=[tuple(result[2][0][0])[0] for result in results]
    movie2=[tuple(result[2][0][1])[0] for result in results]
    support=[result[1] for result in results]
    confidence=[result[2][0][2] for result in results]
    lift=[result[2][0][3] for result in results]
    return list(zip(movie1,movie2,support,confidence,lift))
resultsdf=pd.DataFrame(inspect(results),columns=['Movie 1','Movie 2','Support','Confidence','Lift'])
```
